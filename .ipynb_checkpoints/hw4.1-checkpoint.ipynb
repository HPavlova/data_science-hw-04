```python
from scipy import constants, integrate
from scipy.interpolate import interp1d
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0, 11, 12)
more_time = np.linspace(0, 11, 10000)
time
```

```python
plt.rcParams['figure.figsize'] =(16,10)
plt.plot(time,speed,label='path',linestyle='--',linewidth=2,marker='o',color='g')
plt.title('Chart I', fontsize=20, x=0.5, y=1.1, color='g')
plt.xlabel('Time, hour', fontsize=14, color='g')
plt.ylabel('Speed, km/h', fontsize=14, color='g')
plt.grid()
plt.show()
```

```python
f = interp1d(time, speed, kind='cubic')

plt.rcParams['figure.figsize'] =(16,10)
plt.plot(more_time, f(more_time),label='path',linestyle='--',linewidth=2,marker='o',color='g')
plt.title('Chart II', fontsize=20, x=0.5, y=1.1, color='g')
plt.xlabel('Time, hour', fontsize=14, color='g')
plt.ylabel('Speed, km/h', fontsize=14, color='g')
plt.grid()
plt.show()
```

```python
integrate.quad(f,0,11)
```

```python
f = interp1d(time, speed, kind='quadratic')

plt.rcParams['figure.figsize'] =(16,10)
plt.plot(more_time, f(more_time),label='path',linestyle='--',linewidth=2,marker='o',color='g')
plt.title('Chart III', fontsize=20, x=0.5, y=1.1, color='g')
plt.xlabel('Time, hour', fontsize=14, color='g')
plt.ylabel('Speed, km/h', fontsize=14, color='g')
plt.grid()
plt.show()
```

```python
integrate.quad(f,0,11)
```
